```python
all_data = pd.read_csv('F:/Practicing/Pandas-Data-Science-Tasks-master/SalesAnalysis/All_data.csv')

all_data.head(10)
```

```python
all_data['Months'] = all_data['Order Date'].str[0:2]

all_data.head()
```

```python
all_data.isnull().sum()
```

```python
all_data = all_data.dropna()

all_data.isnull().sum()
```

```python
# converting month column into int type
# the Month column as a type mistake value 'Or', which prevent to from converting to int

#so we should replace it by any int value
all_data['Months'] = all_data['Months'].replace('Or', 0)

all_data['Months'] = all_data['Months'].astype('int32')
all_data.head()
```

```python
# convert non numeric into NaN

# all_data['Months'] = pd.to_numeric(all_data['Months'], errors = 'coerce')
# all_data.Months = all_data.Months.dropna()
#
# all_data.isnull().sum()

all_data.head(10)
```

```python
print(all_data['Quantity Ordered'].iloc[517])
all_data = all_data[all_data['Quantity Ordered'] != 'Quantity Ordered']

all_data = all_data[all_data['Price Each'] != 'Price Each']
```

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])

all_data.head(10)
```

#### Q : what was the best month for sale, how much was earned that month?

```python


# adding sales columns
all_data['Sales'] = (all_data['Quantity Ordered'] * all_data['Price Each'])

all_data.head(10)
```

```python
result = all_data.groupby('Months').sum()
result
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,6))

months = range(1,13)
plt.plot(months, result['Sales'], '.-')

plt.ylabel('Sales Prices in USD', fontsize = 18)
plt.xlabel('Months', fontsize = 18)
plt.yticks(result['Sales'])
plt.xticks(months)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (8,9))

months = range(1,13)
plt.bar(months, result['Sales'])

plt.xticks(months)
plt.yticks( result['Sales'])

plt.tight_layout()
plt.show()
```

#### Q: which city has the highest number of sales

```python
# Qs: which city has the highest number of sales

all_data
```

```python
# Adding city Column

all_data['Cities'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data.head(10)
```

```python
all_data.groupby('Cities')['Sales'].sum()
```

```python
city_sales = all_data.groupby('Cities')['Sales'].sum()
max_city = city_sales.idxmax()
max_sales = city_sales.max()

print(f'the city with highest sales is {max_city}, the Sale is {max_sales}')
```

```python
max_city = all_data.groupby('Cities').sum()

max_city
```

```python
plt.figure(figsize = (8,8))

cities = [city for city, df in all_data.groupby('Cities')] # to get all the Cities in same sequence as Sales

plt.bar(cities, max_city['Sales'])

plt.yticks(max_city['Sales'], size = 8)
plt.xticks(cities, rotation = 20, size = 8)

plt.xlabel('Cities', color = 'm', size = 16)
plt.ylabel('Sales in USD', color = 'm', size = 16)
plt.title('Cities Sales', color = 'm', fontsize = 18)

plt.tight_layout()
plt.show()
```
